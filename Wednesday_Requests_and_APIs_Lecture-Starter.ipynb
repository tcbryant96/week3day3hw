```python

my_api_key = "cedfee00f13f4345a51192228222707"
base_url =  'http://api.weatherapi.com/v1'
api_method = '/current.json'
city = 'Chicago'

api_url = f'{base_url}{api_method}?key={my_api_key}&q={city}'
res = requests.get(api_url)
print(api_url)
weather_data = res.json()
weather_data
```

```python
class weatherAPI:
    def __init__(self,api_key):
        self.api_key = api_key
        self.base_url = 'http://api.weatherapi.com/v1'
    def _get(self, q, api_method, **kwargs):
        request_url = f'{self.base_url}{api_method}?key={self.api_key}&q={q}'
        
        weather_response = requests.get(request_url)
        if weather_response.ok:
            return weather_response.json()
        else:
            print('There was an error')
        
    
    def get_current_weather(self, city_name):
        weather_data = self._get(city_name, '/current.json')
        print(weather_data)
    
    def get_forecast_weather(self, city_name, days = 3):
        weather_data = self.get(city_name, '/forecast.json', days=days)
        
client = weatherAPI("cedfee00f13f4345a51192228222707")
client.get_forecast_weather('Austin')
```

#### Pokemon API Wrapper

Create a python wrapper for the Pokemon API. It should take in a pokemon name and display the pokemon with its height and weight

```python
class pokemonAPI:
    def __init__(self):
        self.base_url = "https://pokeapi.co/api/v2/"
        
    def _get(self, q= ''):
        res = requests.get(f'{self.base_url}pokemon/{q}/')
        if res.ok:
            return res.json()
        else:
            print("There was an error")
            
    def get_height_and_weight(self, poke_name):
        data = self._get(poke_name)
        display_name= f"{poke_name}"
        height = data['height']    
        weight = data['weight']
        print(f"{display_name.title()}'s height is {height} and their weight is {weight}!" )
        

pokemon = pokemonAPI()   
pokemon.get_height_and_weight('meowth')
```
